```
import kagglehub

# Download latest version
path = kagglehub.dataset_download("athiramolps/ice-cream-sales-and-temperature")

print("Path to dataset files:", path)
```

```
import pandas as pd
```

```
import os

# List files in the downloaded path
files_in_path = os.listdir(path)
print("Files in dataset path:", files_in_path)

# Assuming the CSV file is named 'ice_cream_sales.csv' based on common dataset structures or by inspecting the output of files_in_path
# If the exact name is unknown, the user might need to specify or we can infer from the list
csv_file_name = "Ice Cream Sales and Temperature.csv" # Based on typical Kaggle dataset naming and looking for a CSV

# Construct the full path to the CSV file
csv_file_path = os.path.join(path, csv_file_name)

# Read the CSV into a pandas DataFrame
df = pd.read_csv(csv_file_path)

print("DataFrame created successfully. First 5 rows:")
print(df.head())
```

```
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Define independent variable (X) and dependent variable (y)
# X will be 'Temperature (Celsius)' and y will be 'Ice Cream Sales'
X = df[['Temperature (Celsius)']].values # .values converts to numpy array, and [[]] ensures it's 2D
y = df['Ice Cream Sales'].values

# Create a Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Get the coefficients and intercept
coefficient = model.coef_[0]
intercept = model.intercept_

print(f"Linear Regression Model:")
print(f"  Coefficient (slope): {coefficient:.2f}")
print(f"  Intercept: {intercept:.2f}")

# Make predictions
y_pred = model.predict(X)

# Visualize the results
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Temperature (Celsius)'], y=df['Ice Cream Sales'], label='Actual Data')
plt.plot(df['Temperature (Celsius)'], y_pred, color='red', label='Regression Line')
plt.title('Ice Cream Sales vs. Temperature with Linear Regression')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Ice Cream Sales')
plt.legend()
plt.grid(True)
plt.show()
```
